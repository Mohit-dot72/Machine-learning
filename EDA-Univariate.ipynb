```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('titanicDataSet.zip')
```

```python
df.head()
```

1. Catogorical Data

a. Countplot

```python
sns.countplot(df['Survived'])
df['Survived'].value_counts().plot(kind='bar')
```

b. PieChart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

 2. Numerical Data


a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins = 50)
```

b. Displot

```python
sns.distplot(df['Age'])
#sns.kdeplot(df['Age'])
```

c. Boxplot

```python
sns.boxplot(df['Age'])
```
