```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Camparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green')
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'black')

plt.title('Rohit Sharma Vs Virat Kohli Career Camparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color = 'green', linestyle = 'dotted')
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'black',linestyle = 'dotted')

plt.title('Rohit Sharma Vs Virat Kohli Career Camparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], linestyle = 'solid',linewidth=3,marker = 'D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle = 'dashdot',linewidth=3,marker='D',markersize=10)

plt.title('Rohit Sharma Vs Virat Kohli Career Camparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], linestyle = 'solid',linewidth=3,marker = 'D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle = 'dashdot',linewidth=3,marker='D',markersize=10,label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Camparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
```
